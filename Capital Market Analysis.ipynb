```python
df = pd.read_excel(r"E:\Besant Tech\Kaggle\8_BSE NSE\data.xlsx")
df
```

```python
symbol = ((np.asarray(df['Date'])).reshape(12,5))
perchange = ((np.asarray(df['Change in Per'])).reshape(12,5))
```

```python
symbol
```

```python
perchange
```

```python
result = df.pivot(index='Month',columns='SIP Date', values='Change in Per')
result
```

```python
labels = (np.asarray(["{0} \n {1:.2f}".format(symb,value)
                      for symb, value in zip(symbol.flatten(),
                                               perchange.flatten())])
         ).reshape(12,5)
```

```python
fig, ax = plt.subplots(figsize=(12,9))
title = "BSE Sensex Trend on Key SIP Dates for the month of 2022"
plt.title(title,fontsize=18, y=3.05)
ttl = ax.title
ttl.set_position([0.5,1.05])
ax.set_xticks([])
ax.set_yticks([])
sns.heatmap(result,annot=labels,fmt="",cmap='Spectral',linewidths=0.30,ax=ax, vmax=3, vmin=-3)
plt.tick_params(right=False, labelright=False, left=True, labelleft=True, top=True, labeltop=True, labelbottom=False, bottom=False, rotation=90)
plt.show()
```

```python
df = pd.read_excel(r"E:\Besant Tech\Kaggle\8_BSE NSE\Buss_days_sum.xlsx")
df
```

```python
month = df["Month"]
month
```

```python
Positive_days = df["Positive_days"]
Positive_days
```

```python
Negative_days = df["Negative_days"]
Negative_days
```

```python
def addlabels(month,Positive_days, Negative_days):
    total_days = Positive_days + Negative_days
    for i in range(len(month)):
        plt.text(Positive_days[i]/2, i, "{:.2%}".format(Positive_days[i]/total_days[i]), ha="center", color="white", fontweight="bold")
        plt.text(Positive_days[i] + Negative_days[i]/2, i, "{:.2%}".format(Negative_days[i]/total_days[i]), ha="center", color="white", fontweight="bold")

fig = plt.figure(figsize = (15,10))
plt.barh(month, Positive_days, color='green', height=0.5, label="No of Positive Business day")
plt.barh(month, Negative_days, left=Positive_days, color='red', height=0.5, label="No of Negative Business day")
plt.ylabel("Months")
plt.xlabel("Count of Days")
plt.title("No of business days in 2022")
plt.legend()

addlabels(month, Positive_days, Negative_days)
plt.show()
```
